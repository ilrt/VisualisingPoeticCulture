Load all the sheets as a single data frame:

```python
import pandas as pd
import visualising_poetry as vp
vp.VisualisingPoetrySetup().setup_if_needed()
df = vp.VisualisingPoetryQuery().pickle_as_single_data_frame()
```

Total number of poems per year:

```python
# number of poems by year
total_poems_by_year = df.groupby('year')['first line'].count()
print(total_poems_by_year)
total_poems_by_year.plot()

```

Total number of poems per year by publication:

```python
# number of poems by year and publication
total_poems_by_year_pub = df.groupby(['year', 'publication title'])['first line'].count()
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
    print(total_poems_by_year_pub)
    
total_poems_by_year_pub.plot()
```
